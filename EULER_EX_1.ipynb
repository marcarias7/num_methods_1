                          Movimiento de una esféra en un medio viscoso





```python
#IMPORTAMOS LAS LIBRERÍAS
import matplotlib.pyplot as grf
import numpy as np
```

```python
#ESTABLECEMOS LAS CONDICIONES INICIALES 
to = 0
xo = 0
vo = 0
n = 100
tf = 1.5
h = (tf - to)/n
print("incremento_tiempo=",round(h,4))
```

```python
def dvdt(t,v):
    return 9.0 - 40.0*v

t = np.zeros(n+1)
t[0]=to
for i in range(n):
    t[i+1] = to + h*i
    
#METODO DE EULER    
v = np.zeros(n+1)
v[0]=vo
for i in range(n):
    v[i+1] = v[i] + h*dvdt(t[i],v[i])
print("velocidad_limite =",round(v[i+1],5))

x = np.zeros(n+1)
x[0] = xo
for i in range(n):
    x[i+1] = x[i] + h*v[i]
print("en tf =",tf,"segundos, posición_aproximada =",round(x[i+1],5))
```

```python
from math import log , e
grf.title("dv/dt = 9.0 - 40.0v , v(0) = 0")
grf.plot(t,v,"o-",label="solución aproximada")
grf.xlabel("t")
grf.ylabel("v")
grf.ylim(0.0,0.3)
grf.xlim(-0.2,0.4)
def f_v(t,v):
    return (9.0/40.0)*(1 - e**(-40*t))
grf.plot(t,f_v(t,v),"r-",label="solución exacta")
grf.legend(loc="upper left")
grf.grid()
```
