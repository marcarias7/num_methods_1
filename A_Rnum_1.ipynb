```python
import numpy as np
import matplotlib.pyplot as grf
from math import*
n = 50
to = 0.0
vo = 0.0
tf = 1.5
v = np.zeros(n+1)
t = np.zeros(n+1)
h = (tf-to)/n 
for i in range(n+1):
    t[i] = round(to + i*h,7)  
def f_rk(t,v):
    return 9.0 - 40.0*v
v[0] = vo
for i in range(n):
    k1 = h*f_rk(t[i],v[i])
    k2 = h*f_rk(t[i] + h/2,v[i] + k1/2)
    k3 = h*f_rk(t[i] + h/2,v[i] + k2/2)
    k4 = h*f_rk(t[i] + h,v[i] + k3)
    v[i+1] = v[i] + (k1 + 2*k2 +2*k3 + k4)/6.0    
grf.title('RK4_dv/dt = 9 - 40v , v(0) = 0 , n = 50')
grf.plot(t,v,"o-",color ='blue',label="aproximada")
t = np.linspace(0,5,1000)
grf.plot(t,9*(1 - e**(-40.0*t))/40,color = 'red',label="exacta")
grf.xlabel("EJE_t")
grf.ylabel("EJE_v(t)")
grf.grid(True)
grf.legend(loc="upper left")
grf.ylim(0,0.3)
grf.xlim(-0.2,0.5)
```
